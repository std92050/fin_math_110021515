# Mean reversion strategy

## Pair trading
We choose Coca cola(price $p_1$) and Pepsico(price $p_2$) as the target pair. 

Assumption: both $p_1$ and $p_2$ are $I(1)$, i.e $\{p_i(t) - p_i(t-1)\}_{t=1}$ is stationary.

Then these two are said to be **cointegrate** if existing some linear combination process of $p_1$ and $p_2$ that is weak stationary.


```python
tickers = ['PEP','KO']
data =yf.download(tickers,start='2015-01-01', end ='2025-01-01')
data.head()
```

### Constructing spread

```python
X = sm.add_constant(data_2023['PEP'].values)
ls = sm.OLS(data_2023['KO'].values,X).fit()
```

```python
ls.params
```

One can also use scikit-learn to do the linear regression, which gives the same result

```python
import sklearn as skl
from sklearn.linear_model import LinearRegression

reg=skl.linear_model.LinearRegression().fit( data_2023['PEP'].values.reshape(-1, 1),data_2023['KO'].values )

print(f'estimated slope is {reg.coef_[0]}')
```

From least square we get the estimated linear combination coefficient $\beta =  0.29752124$ and we can define the spread

$$
 S(t) = p_1(t) - \beta p_2(t)
$$

```python
b = 0.29752124
spread =  data_2023["KO"] - b*  data_2023['PEP']
plt.plot(spread)
```

[Augmented Dickey–Fuller test](https://en.wikipedia.org/wiki/Augmented_Dickey%E2%80%93Fuller_test) shows the weak stationarity of spread is statistically significant.

```python
from statsmodels.tsa.stattools import adfuller
adf_test = adfuller(spread)
print(f'p-value is {adf_test[1]}')
```

### Construct long/short strategy
First we can try $\mu \pm \sigma$ as upper(U) and lower(L) thresh hold

```python
mu = spread.mean()
s = spread.std()
L= mu -s
U= mu +s
plt.plot(spread)
plt.axhline(y = mu , ls='--',c='r')
plt.axhline(y =L , ls='--',c='r')
plt.axhline(y = U, ls='--',c='r')
```

The idea of trading strategy is that:

* Each time the spread crosses the mean value from outside the band, we liquidate the spread position.
* The first time the spread crosses the upper level from inside the band after leaving mean value, we long one unit of KO and short $\beta$ unit of PEP.
* The first time the spread crosses the lower level from inside the band after leaving mean value, we short one unit of KO and long $\beta$ unit of PEP.

Here since we consider the price difference as the spread, each time we trade constant shares in long and short position.


We can also formulate these trading timing following this [paper](https://arxiv.org/abs/2011.02870). Let $\tau^-_0= \theta^-_i=\tau^+_0= \theta^+_i=0$
and 

\begin{align*}
\forall i\geq 1,\quad &\tau^+_i=\inf\{ t>\theta^+_{i-1}\mid S(t)\geq S_U \} \quad 
\theta^+_i=\inf\{ t>\tau_{i}^+ \mid  S(t)\leq \mu \} 
\\\\
&\tau^-_i=\inf\{ t>\theta^-_{i-1} \mid S(t)\leq S_L \} \quad \theta^-_i=\inf\{ t>\tau^-_{i}\mid S(t)\geq \mu \} 
\end{align*}

In our setting, $S_U = \mu + \sigma$ and $S_L = \mu - \sigma$. 

```python
data_2023['spread'] = spread
```

```python
spread.values
```

```python
signal = np.zeros_like(spread.values)

for t in range(len(spread)):
    #initial trading signal setting
    if t== 0:
        if spread.iloc[t] > U:
            signal[t] = -1
        elif spread.iloc[t]<L:
            signal[t]=1
        else:
            signal[t]=0
    #for t>0, trading signal depends on previous state
    else:
        if  signal[t-1] ==1:
            if  spread.iloc[t] < mu:
                signal[t]= 1
            elif spread.iloc[t]>U:
                signal[t]=-1
            else:
                signal[t] =0
                
        elif  signal[t-1] ==0:
            if  spread.iloc[t] < L:
                signal[t]= 1
            elif spread.iloc[t]>U:
                signal[t]=-1
            else:
                signal[t] =0
                
        else:
            if  spread.iloc[t] > mu:
                signal[t]= -1
            elif spread.iloc[t]<L:
                signal[t]=1
            else:
                signal[t] =0
                
```

```python
data_2023['signal'] = signal
```

We can visualize our strategy on time series plot of the spread. Black dots correspond to flat positions, red dots correspond to short positions and blue dots correspond to long positions. Totally 15 trades in 8 years.

```python
long=data_2023[data_2023['signal']==-1].index
short=data_2023[data_2023['signal']==1].index
none=data_2023[data_2023['signal']==0].index
plt.figure(figsize=(12,8))
plt.scatter( spread[long].index,spread[long], c='r',s=5 )
plt.scatter( spread[short].index,spread[short], c='b',s=5 )
plt.scatter( spread[none].index,spread[none], c='k',s=5 )
plt.plot(spread,lw=0.5)
plt.axhline(y = mu , ls='--',c='g')
plt.axhline(y =L , ls='--',c='g')
plt.axhline(y = U, ls='--',c='g')
```

### Backtest
Since we are not self-financed, we also consider a cash position with one dollar as initial capital. We compute the cumulative strategy return at time $t$ as
$$
R_{\text{strategy}}(t) = 
\frac{\text{cash at time t} \; + \; \text{pair position value} \; - \; \text{initial capital}}{\text{initial capital}} = \frac{\text{cash at time t} \; + \; \tilde{S}(t)-1}{1}
$$

where $\tilde{S}(t) = sig(t) S(t)$ and $sig(t)$ is the signal at time $t$.

```python
data_2023.loc[:,'cash'] = 1
data_2023.loc[:,'position value'] = data_2023['signal']*spread
data_2023.loc[:,'cumR']= 0

data_2023['cash'] = data_2023['cash'].astype(float)
data_2023['cumR'] = data_2023['cumR'].astype(float)

for i in range(len(data_2023)):
    if i==0:
        continue
    t_n = data_2023.index[i]
    t_o = data_2023.index[i-1]
    if data_2023.loc[t_o,'signal'] != data_2023.loc[t_n,'signal']:
        diff= data_2023.loc[t_n,'position value'] - data_2023.loc[t_o,'position value']
        data_2023.loc[t_n,'cash'] = data_2023.loc[t_o,'cash'] - diff
    else:
        data_2023.loc[t_n,'cash'] = data_2023.loc[t_o,'cash']
    data_2023.loc[t_n,'cumR'] =  data_2023.loc[t_n,'cash'] + data_2023.loc[t_n,'position value'] -1
    
```

```python
plt.plot(data_2023['cumR'])
```
