```python
#remove
ETF_50=ETF_50.dropna()
```

```python
ETF_50.keys()
```

```python
price = ETF_50[["開盤價","收盤價"]]
price
```

## Price path from data

First check the length of the data 

```python
days = len(price.index)
days
```

Extracting the price to an array 

```python
price_array = np.empty(0)
for t in range(days) :
    price_in_a_day = [price.loc[t , "開盤價" ] ,price.loc[t ,  "收盤價"] ]
    price_array = np.append(price_array , price_in_a_day)
```

```python
plt.plot(price_array)
```
