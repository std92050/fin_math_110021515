```python
# a dictionary that map ticker to its underlying stock
name_dic ={}
for s in majorStockIdx.Symbol :
    name_dic[s] = yf.Ticker(s).info.get('shortName')
```

#### Load each index 
Collect historical stock data for world major stock indices from January 2010 until 30 September 2020 and concatenate them into a big df.

```python

stock_list = []
for s in majorStockIdx.Symbol: # iterate for every stock indices
    # Retrieve data from Yahoo! Finance
    tickerData = yf.Ticker(s)
    tickerDf1 = tickerData.history(period='1d', start='2010-1-1', end='2024-2-10')
    # Save historical data 
    tickerDf1['ticker'] = s # don't forget to specify the index
    stock_list.append(tickerDf1)
# Concatenate all data
msi = pd.concat(stock_list, axis = 0)
```

Dictionary for classification of indicies w.r.t region.

No ^CASE30 data ! Also we drop ^JN0U.JO and ^IMOEX.ME data since they don't start from 2010.

```python
msi = msi[ (msi.ticker != '^JN0U.JO') & (msi.ticker !='IMOEX.ME' )]
```

```python
region_idx= { 'US & Canada' : ['^GSPC', '^DJI', '^IXIC', '^RUT','^GSPTSE'],
  'Latin America' : ['^BVSP', '^MXX', '^IPSA'],
  'East Asia' : ['^N225', '^HSI', '000001.SS', '399001.SZ', '^TWII', '^KS11'],
  'ASEAN & Oceania' : ['^STI', '^JKSE', '^KLSE','^AXJO',  '^NZ50'],
  'South & West Asia' : ['^BSESN', '^TA125.TA'],
  'Europe' : ['^FTSE', '^GDAXI', '^FCHI', '^STOXX50E','^N100', '^BFX']
}
```

```python
# use the dic to add another col 'region'
def getRegion(ticker):
    for k in region_idx.keys():
        if ticker in region_idx[k]:
            return k
msi['region']= msi.ticker.apply(lambda x: getRegion(x))
```

Compute the growth/return of each indecies w.r.t their start time. We add a new column 'chBegin' to store the result.

```python
tempt = msi.groupby('ticker',sort = False).apply(lambda x: (x.Close/x.Close[0] - 1) * 100)
msi['chBegin']=tempt.droplevel('ticker')
```

```python
# some setting on datetime index of msi (unify expression and then keep date only)
msi=msi.reset_index()
msi.Date = msi.Date.apply(lambda x : (x.tz_convert('UTC')).date() )
msi.Date = pd.to_datetime(msi.Date)
msi.set_index('Date',inplace = True)
```

Transform the dataset into tickers as columns, and rows as the price change for each date.

```python
# Transform the data to be ticker column-wise

chBegin = msi.groupby(['Date', 'ticker'])['chBegin'].first().unstack() # here first() or last() are the same, only one value for each ticker at a time

# Fill null values with the values on the row before
chBegin = chBegin.fillna(method='bfill')
```

```python
chBegin.keys()
```

```python

fig, axes = plt.subplots(3,2, figsize=(12, 8),sharex=True)
pagoda = ["#965757", "#D67469", "#4E5A44", "#A1B482", '#EFE482', "#99BFCF"] # for coloring
for i, k in enumerate(region_idx.keys()):
# Iterate for each region
    ax = axes[int(i/2), int(i%2)]
    for j,t in enumerate(region_idx[k]):
        # Iterate and plot for each stock index in this region
        ax.plot(chBegin.index, chBegin[t], marker='', linewidth=1, color = pagoda[j])
        ax.set_title(k, fontweight='bold')
    ax.legend( [ name_dic[t]  for t in region_idx[k]], loc='upper left', fontsize=7)

fig.text(0.5,0, "Year", ha="center", va="center", fontweight ="bold")
fig.text(0,0.5, "Price Change/Return (%)", ha="center", va="center", rotation=90, fontweight ="bold")
fig.suptitle("Price Change/Return for Major Stock Indices based on 2010", fontweight ="bold",y=1.05, fontsize=14)
fig.tight_layout()
```

### Resampling
Here we try to resample the data with different frequency so that the date of each data point is the 30th. But it's possible that such date is not business day. One way to fix this is to choose the last business day before the 30th.

```python
def nearest(dates, dateRef):
    # Get the previous days before the reference date (dateRef)
    prevDate = dates[dates < dateRef]
    return prevDate[-1] # return the last date
    
def getReturn(period, number, ticker, dt, val):
    # Subset the dataset by stock index
    df = msi[msi.ticker == ticker].reset_index()
    # Get all business days
    
    #existingDates = df['Date'].unique() This will have ordering problem, not depend on the order of date but the order on msi.index
    existingDates = np.unique( df.Date)
    # Get the past date on months or some periods before
    if period == 'Y':
        dtp = (pd.Timestamp(dt) - pd.DateOffset(years=number)) #dateoffset can deal with business day, leap year etc, which timedealta can't do.
    elif period == 'M':
        dtp = (pd.Timestamp(dt) - pd.DateOffset(months=number)) #.to_numpy()
    elif period == 'W':
        dtp = (pd.Timestamp(dt) - pd.DateOffset(weeks=number)) #.to_numpy()
    elif period == 'D':
        dtp = (pd.Timestamp(dt) - pd.DateOffset(days=number)) #.to_numpy()
        
    # check if the past Date was business day
    # if not, get the closest business day
    # then, return the price change
    
    if dtp in existingDates: 
        return (val/df.loc[df.Date == dtp, "Close"].values[0] - 1)*100
    else:
        closestDate = nearest(existingDates, dtp)
        return(val/ df.loc[df.Date == closestDate, "Close"].values[0] - 1)*100
```

```python
lastDate = msi[msi.index == '2020-09-30'].reset_index()

lastDate['6MR'] = lastDate.apply(lambda r: getReturn('M', 6, r.ticker, r.Date, r.Close), axis =1) #apply on row
lastDate['1YR'] = lastDate.apply(lambda r: getReturn('Y', 1, r.ticker, r.Date, r.Close), axis =1)
lastDate['3YR'] = lastDate.apply(lambda r: getReturn('Y', 3, r.ticker, r.Date, r.Close), axis =1)
lastDate['5YR'] = lastDate.apply(lambda r: getReturn('Y', 5, r.ticker, r.Date, r.Close), axis =1)
lastDate['10YR'] = lastDate.apply(lambda r: getReturn('Y', 10, r.ticker, r.Date, r.Close), axis =1)
```

```python
fig, axes = plt.subplots(1,5, figsize=(20, 10),sharey=True)
width = 0.75
cols = ['6MR','1YR','3YR', '5YR', '10YR']
for i, j in enumerate(cols):
    ax = axes[i]    
    tick = lastDate.ticker.apply(lambda t : name_dic[t])
    ax.barh(tick,lastDate[j], width, color = pagoda[i])
    ax.set_title(j, fontweight ="bold")    
    ax.invert_yaxis()
fig.text(0.5,0, "Return (%)", ha="center", va="center", fontweight ="bold")
fig.text(0,0.5, "Stock Indices", ha="center", va="center", rotation=90, fontweight ="bold")
fig.suptitle("Returns for Major Stock Indices based on 30 September", fontweight ="bold",y=1.03, fontsize=14)
fig.tight_layout()
```
